```r
set.seed(42)
```

```r
x <- 1:100
y <- 3 + 4 * x + rnorm(100, sd = 100)
x
y
rnorm(100,sd=100)
```

```r
simple_model <- lm(y ~ x)
simple_model
complex_model_02 <- lm(y ~ poly(x,2))
complex_model_02
complex_model_10 <- lm(y ~ poly(x,10))
complex_model_10
complex_model_26 <- lm(y ~ poly(x,26))
complex_model_26
```

```r
plot(x, y, main="Overfitting vs Ajuste Simples")
lines(x, predict(simple_model), col = "blue",lwd = 2)
lines(x, predict(complex_model_02), col = "green",lwd = 2)
lines(x, predict(complex_model_10), col = "red",lwd = 2)
lines(x, predict(complex_model_26), col = "purple2",lwd = 2)

legend("bottomright",
       legend = c("Linear",
                  "Polinomial grau 2",
                  "Polinomial grau 10",
                  "Polinomial grau 26"),
      col = c("blue","green","red","purple"),
      lwd = 1,
      cex = 0.5)
```
